```python
import pandas as pd
df = pd.read_csv("insurance.csv")
```

```python
print(df.head())
df.tail()
```

```python
df.isnull().sum()
#all the dataset is fine , so no issues
```

```python
df['Hospital charges']=df['charges']
df
```

```python
df = df.drop(columns=['charges'])

df
```

```python
df.replace({'sex': {'male': 0, 'female': 1}}, inplace=True)
df.replace({'smoker': {'no': 0, 'yes': 1}}, inplace=True)
df.replace({'region':{'southeast':0,'southwest':1,'northeast':2,'northwest':3}}, inplace=True)
df
```

```python
df.describe()
```

```python
x = df.iloc[:,:6]
x

```

```python
y = df.loc[:,['Hospital charges']]
y
```

```python
#training the data taking test-size as 10 percent as we will train the 
#data with more size , better accuracy
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=0)
```

```python
#taking the predicted output using teh algorithms .
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

#object initialise
log = LinearRegression()
dcs = DecisionTreeRegressor()
#fitting the data inside and performing the training
log.fit(x_train,y_train)
dcs.fit(x_train,y_train)

pred1 = log.predict(x_test)
pred2 = dcs.predict(x_test)


print(f"Predict salary for Linear Regression {pred1} & accuracy is {r2_score(y_test,pred1)}")


```

```python
print(f"Predict salary for Decision Tree {pred2} & accuracy is {r2_score(y_test,pred2)}")
```
